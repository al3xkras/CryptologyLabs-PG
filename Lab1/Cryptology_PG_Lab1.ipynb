```python
import io
# -*- encoding: utf-8 -*-
"""
Program czyście tekst ze znaków białych, symboli diakrytycznych, cyfr, 
a litery zamienia na małe oraz formatuje tekst w 7 kolumn po 5 znaków.
"""

def wczytaj(a):
    nazwa_pliku = input("Podaj nazwe pliku tekstowego bez rozszerzenia: \nNazwa: ")
    try:
        t = io.open(nazwa_pliku+ ".txt",mode="r",encoding="utf-8")
        a = t.read()
        t.close()
        return a.lower()
    except:
        print("Niepoprawna nazwa pliku. \n")
        return wczytaj(a)
    
    
def usun_znaki(table,rm_sp):
    print(table)
    sym=" .,'!:;/@#$%^&(*)-_+=[{]}|?\n\""
    if not rm_sp:
        return rm_sym(usun_plznaki(table),sym)

    s1=input("symbols to replace:")
    s2=input("replace with:")
    
    return rm_sym(usun_plznaki(replace_sym(table,s1,s2)),sym)

def rm_sym(table,symbole):
    b=range(len(symbole))
    for i in b:
        table=table.replace(symbole[i],"")
    return(table)

def replace_sym(table,znaki,znaki2):
    b=range(len(znaki))
    for i in b:
        table=table.replace(znaki[i],znaki2[i]) 
    return(table)

def usun_plznaki(table):
    return replace_sym(table,"ąęółżźćń","aeolzzcn")


def usun_cyfry(table):
    cyfry="0123456789"
    b=range(len(cyfry))
    for i in b:
        table=table.replace(cyfry[i],"")
    return(table)     
    
def przeksztalcenie(table,table2):
    size = range(len(table))
    k = 0
    for i in size:
    
        if i%5 == 0 and i !=0 and i%35 !=0:
            k+=1
            table2 = table2+" "
            #table2[i+k]=" "
        if i%35==0 and i != 0:
            k+=1
            table2 = table2+"\n"
            #table2[i+k]="\n"
        table2 = table2 + table[i]
    return(table2)
    
def zapis(table2,print_file):
    if (print_file):
        print(table2)
    nazwa_pliku = input("Podaj nazwe pliku do zapisu: \nNazwa: ")
    f = io.open(nazwa_pliku + ".txt", mode='w',encoding='utf-8')
    f.write(table2)
    f.close()
    
    

table = ""   
table2= "" 

remove_nums=input("Remove numbers? (T/N): ").upper()=="T"
remove_sym=input("Remove special symbols? (T/N): ").upper()=="T"
table_format=input("Format table? (T/N): ").upper()=="T"
to_upper=input("To uppercase? (T/N): ").upper()=="T"
print_file=input("Print file? (T/N): ").upper()=="T"
table = wczytaj(table)

if (remove_sym):
    table = usun_znaki(table,True)
else:
    table = usun_znaki(table,False)

if (remove_nums):
    table = usun_cyfry(table)

if to_upper:
    table = table.upper()
else:
    table = table.lower()
    
if (table_format):
    table2 = przeksztalcenie(table,table2)
else:
    table2=table
    
zapis(table2,print_file)
```
