```python
import os
import io
from random import shuffle

def formatText(remove_nums="T"):
    remove_sym="T"
    table_format="N"
    to_upper="T"
    print_file="N"
    replace_symbols="\"	–ś\""
    replacements="\"  s\""
    f_in="text.txt"
    f_out="text_out.txt"
    args=[remove_nums,remove_sym,table_format,to_upper,print_file,replace_symbols,replacements]
    cmd="python ./readData.py \"%s\" \"%s\" "%(f_in,f_out)+" ".join(args)
    assert os.system(cmd)==0
    text=None
    with io.open(f_out,mode="r",encoding="utf-8") as f:
        text=f.read()
    return text
```

```python
class CaesarEncoding:
    
    def __init__(self,n,alphabet):
        self.n=n
        self.alphabet=[x for x in alphabet]
        self.alp=dict((alphabet[i],i) for i in range(len(alphabet)))
    
    def encode(self,symbol):
        return self.alphabet[(self.alp[symbol]+self.n)%len(self.alphabet)]
    
    def decode(self,symbol):
        return self.alphabet[(self.alp[symbol]-self.n)%len(self.alphabet)]
    
    def decodeString(self,string):
        return "".join(map(self.decode,string))
    
    def encodeString(self,string):
        return "".join(map(self.encode,string))


class Albasz:
    def __init__(self,alphabet):
        self.alphabet=alphabet
        self.alphabet2=alphabet[::-1]
        self.alp=dict((alphabet[i],i) for i in range(len(alphabet)))
        self.alp2=dict((self.alphabet2[i],i) for i in range(len(self.alphabet2)))
    
    def encode(self,symbol):
        return self.alphabet2[self.alp[symbol]]
    
    def decode(self,symbol):
        return self.alphabet[self.alp2[symbol]]
    
    def decodeString(self,string):
        return "".join(map(self.decode,string))
    
    def encodeString(self,string):
        return "".join(map(self.encode,string))


class MonoalphabeticEncoding:
    def __init__(self,alphabet):
        self.alphabet=alphabet
        self.alphabet2=[x for x in alphabet]
        shuffle(self.alphabet2)
        self.alp=dict((alphabet[i],i) for i in range(len(alphabet)))
        self.alp2=dict((self.alphabet2[i],i) for i in range(len(self.alphabet2)))
    
    def encode(self,symbol):
        return self.alphabet2[self.alp[symbol]]
    
    def decode(self,symbol):
        return self.alphabet[self.alp2[symbol]]
    
    def decodeString(self,string):
        return "".join(map(self.decode,string))
    
    def encodeString(self,string):
        return "".join(map(self.encode,string))\


class Affine:
    def __init__(self,a,b, alphabet):
        self.alphabet=alphabet
        self.alp=dict((alphabet[i],i) for i in range(len(alphabet)))
        self.a=a
        self.b=b
        gcd, self.a_inv, _=Affine.gcdExtended(a,len(alphabet))
        assert gcd==1
        if (self.a_inv<0):
            self.a_inv+=len(alphabet)
        
    @staticmethod
    def gcdExtended(a, b):
        if a == 0 :
            return b,0,1
        gcd,x1,y1 = Affine.gcdExtended(b%a, a)
        x = y1 - (b//a) * x1
        y = x1
        return gcd,x,y
        
    def encode(self,symbol):
        return self.alphabet[(self.a*self.alp[symbol]+self.b)%len(self.alphabet)]
    
    def decode(self,symbol):
        return self.alphabet[(self.a_inv*(self.alp[symbol]-self.b))%len(self.alphabet)]
    
    def decodeString(self,string):
        return "".join(map(self.decode,string))
    
    def encodeString(self,string):
        return "".join(map(self.encode,string))
```

```python
alphabet=[chr(x) for x in range(ord('A'),ord('Z')+1)]

rm_num=input("remove numbers? T/N: ")[0].upper()=="T"

text=formatText("T" if rm_num else "N")

if not rm_num:
    alphabet=alphabet+list(map(str,range(10)))
    
encodings=[
    CaesarEncoding(14,alphabet),
    Albasz(alphabet),
    MonoalphabeticEncoding(alphabet),
    Affine(7,3,alphabet)
]
for encoding in encodings:
    print(type(encoding))
    print("Encoded text: "+encoding.encodeString(text))
    assert encoding.decodeString(encoding.encodeString(text))==text
    print()
```
